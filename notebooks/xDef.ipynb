```python
import xgboost as xgb
import pandas as pd
import numpy as np
```

```python
df_int = pd.read_csv("interceptions_players.csv", index_col=0)
```

```python
df_tack = pd.read_csv("tackles_players.csv", index_col=0)
```

```python
df_int
```

```python
df_tack["y_pred_int"] = df_int["y_pred"]
df_tack["y_pred_int"] = df_tack["y_pred_int"].replace([np.nan, -np.inf], 0)

df_tack["expected_y_int"] = df_int["expected_y"]
df_tack["expected_y_int"] = df_tack["expected_y_int"].replace([np.nan, -np.inf], 0)

df_tack["liczba przechwytów"] = df_int["liczba_przechwytów"]
df_tack["liczba przechwytów"] = df_tack["liczba przechwytów"].replace([np.nan, -np.inf], 0).astype(int)

df_tack["y_pred_tack"] = df_tack["y_pred"]
df_tack["expected_y_tack"] = df_tack["expected_y"]
df_tack["xDef_tackle per 90 minutes"] = df_tack["xDef per 90 minutes"]
df_tack["xDef_interception per 90 minutes"] =  (90 * df_tack["y_pred_int"]) / df_tack["minutes"]
```

```python
order = ["nickname","position","club","liczba odbiorów", "odbiory udane", "odbiory nieudane", "y_pred_tack", "expected_y_tack", "liczba przechwytów", "y_pred_int", 
         "expected_y_int", "minutes", "starts", "matches", "xDef_tackle per 90 minutes", "xDef_interception per 90 minutes"]
df_merge = df_tack[order]
df_merge
```

```python
df_merge["xDef"] = df_merge["y_pred_tack"] + df_merge["y_pred_int"]
df_merge["xDef per 90 minutes"] = (90 * df_merge["xDef"]) / df_merge["minutes"]
```

```python
df_merge = df_merge.sort_values(by='xDef', ascending=False)
df_merge[:15]
```

```python
df_xDef = df_merge[["nickname","position","club","liczba odbiorów", "odbiory udane", "odbiory nieudane", "liczba przechwytów",  "minutes", "starts", "matches", "xDef", "xDef per 90 minutes"]]
```
